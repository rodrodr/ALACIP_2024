


## Erosión democrática y comportamiento parlamentario

<br>

¿Cómo investigar procesos de erosión democrática a partir del comportamiento parlamentario?

<br>

¿Qué tipos de información y estrategias metodológicas disponemos para entender el posicionamiento y las acciones de los legisladores frente a la democracia?


## ¿Qué dice la literatura?

<br>

Estudios con foco en el **ejecutivo**:

<br>

1. **Control** político del legislativo (supermayorías, etc.)

2. **Poder de agenda** del ejecutivo

3. Inestabilidad de **coaliciones** de gobierno

::: {.notes}
Poder de agenda del ejecutivo incluye la capacidad de proponer y aprobar leyes, así como de vetarlas y establecer preferencias en la tramitación de las leyes. De ese modo, el legislativo tiene que reaccionar a la iniciativa del ejecutivo. 
::: 



## ¿Qué dice la literatura?

<br>

Estudios con foco en el **legislativo**:

<br>

1. **Polarización** legislativa (votos y debates)

2. **Oposición** y capacidad de resistencia al ejecutivo

3. Abuso del **poder de "oversight"** (control)

4. **Recorte de derechos** fundamentales por ley

::: {.notes}
Polarización - falta de capacidad de colaboración y llegar a consensos en la aprobación de leyes.

Poder "oversight" - procesos de impeachment.

Recorte de derechos fundamentales por ley - leyes iniciadas en el mismo legislativo que restringen derechos civiles y políticos.
::: 



## ¿Qué métodos emplea?

<br>

1. Análisis de contenido de debates legislativos

2. Tamaño y composición de bloques partidarios
(gobierno y oposición)

3. Redes y patrones de votación y coautoría

4. Escalado de posiciones
(polarización y diversidad ideológica)

4. Análisis de procesos de control

## ¿Con qué datos?

<br>

- Características institucionales

- Intervenciones en pleno

- Composición de las coaliciones (gobierno y oposición)

- Producción y éxito legislativos

- Votaciones nominales (alineamientos, disciplina)

- Procedimientos de control y fiscalización


## PELA Comportamiento

<br>

Extensión del Proyecto PELA-USAL (Opinión)

<br>

**Objetivo:** crear una infraestructura abierta y comparable de datos de **comportamiento** legislativo

<br>

**Países:** 8 países de América Latina (posibilidad de 12)

<br>

**Tiempo:** varía según la disponibilidad de datos

(hasta 7 legislaturas en algunos casos)

::: {.notes}
Países: AR, BR, CL, CO, EC, MX, PE, PY

Países a incluir: CR, RD, PA, UY
::: 



## PELA Comportamiento

<br>

Datos sobre comportamiento legislativo recolectados:

- intervenciones de los diputados en las sesiones

- asistencia y votaciones nominales

- comisiones y sus integrantes

- iniciativas legislativas, sus autores y tramitación

- iniciativas destinadas a cada comisión

- partidos y cambios de partido



## Diagrama de flujo del proyecto

[![](Data_Pipeline.png)](Data_Pipeline.png)

## Conceptos de trabajo

<br>

Modularidad 🧱

<br>

Interfacialidad 🔌

<br>

Diferentes perfiles de uso y usuario 💻 🖥️📱👩‍🎓👨‍💼👩‍💻🤖🎓📈📚

<br>

Flexibilidad horizontal y vertical:

-   niveles de análisis

-   configuraciones

## Arquitectura: Capas conceptuales

::: columns
::: {.column width="40%"}
![](PELA_Layers.png)
:::

::: {.column width="60%"}
Capas: espacios de organización conceptual y técnica

<br>

Jerarquizadas en función de una lógica técnico-analítica

<br>

No obstante, superpuestas e interdependientes

<br>

Etapas del proyecto
:::
:::

## Arquitectura: Capa "Datos brutos"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Bruto.png)
:::

::: {.column width="60%"}
Datos originales de cada país en diferentes niveles de detalle

<br>

Formatos distintos (html, pdf, xml, json, csv, etc.)

<br>

No estandarizados

<br>

No aptos para el análisis

(mayoría de los países)
:::
:::

## Arquitectura: Capa "Estandarizados"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Estandar.png)
:::

::: {.column width="60%" align="center"}
![](data_model.png)
:::
:::

## Arquitectura: Capa "Estandarizados"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Estandar.png)
:::

::: {.column width="60%"}
Modelo de datos estandarizados

<br>

Comparabilidad máxima: entre países y entre períodos legislativos

<br>

Estructural relacional (SQL)

<br>

Modular para permitir flexibilidad

<br>

Usuarios avanzados
:::
:::

## Arquitectura: Capa "Indicadores"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Indicadores.png)
:::

::: {.column width="60%"}
Indicadores derivados de la literatura (W-Nominate, Rice, etc.)

<br>

Y derivados de la investigación
(clasificación temática, etc.)

<br>

Facilitar el análisis y ampliar el abanico de usuarios

<br>

Todos los tipos de usuario
:::
:::

## Arquitectura: Capa "Vistas"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Vistas.png)
:::

::: {.column width="60%"}
Perspectivas o representaciones sintéticas o simplificadas de la base de datos

<br>

Facilitan la consulta de datos

<br>

Adecuación a perfiles de uso

<br>

Ejemplos: intervenciones en cada sesión, votos por partido, redes de coautoría, etc.
:::
:::

## Arquitectura: Capa "Interfaces"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Interfaces.png)
:::

::: {.column width="60%"}
Conexiones con otras bases de datos

<br>

API y documentación en inglés, español y portugués

<br>

Paquetes en R y Python

<br>

Gráficos interactivos y tablas para usuarios no expertos
:::
:::

## Arquitectura: Capa "Usuarios"

::: columns
::: {.column width="40%"}
![](ALACIP_2024_Layer_Usuarios.png)
:::

::: {.column width="60%"}
Adaptación a diversos perfiles de usuario potencial:

-   Usuarios con sólidos conocimientos de ciencia de datos

-   Académicos con fluencia en paquetes estadísticos

-   Académicos no expertos en estadística

-   Medios de comunicación y organizaciones de la sociedad civil
:::
:::

## Desarrollo del proyecto

<br>

**Etapa 1:** descarga de datos originales [✓ (posib. nuevos casos)]{style="color:green"}

<br>

**Etapa 2:** estandarización [(en proceso)]{style="color:orange"}

<br>

**Etapa 3:** indicadores y vistas [(no iniciada)]{style="color:red"}

<br>

**Etapa 4:** interfaces [(no iniciada)]{style="color:red"}

<br>

**Etapa 5:** publicación de los datos [(no iniciada)]{style="color:red"}
